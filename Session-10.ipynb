```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.normal(0,1,1000)


width = 1
histmin = np.floor(min(rng))
histmax = np.ceil(max(rng))+width

bins = np.arange(histmin, histmax, width)

plt.hist(rng, bins= 100, alpha=0.5, edgecolor = 'black')
plt.xlabel('random number')
plt.ylabel('number of times chosen')
plt.savefig('session_10.pdf', bbox_inches='tight', facecolor='orange')
```
